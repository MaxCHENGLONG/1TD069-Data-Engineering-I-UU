```python
import pandas as pd
import matplotlib.pyplot as plt

def read_and_process_data(file_path):
    data = pd.read_csv(file_path, sep=',', header=None, names=['Pronoun', 'Count'])
    
    total_count = data['Count'].sum()
    
    data['Percentage'] = (data['Count'] / total_count) * 100
    
    return data

def plot_pronoun_percentages(data):
    # Plotting the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(data['Pronoun'], data['Percentage'], color='skyblue')
    plt.xlabel('Pronoun')
    plt.ylabel('Percentage (%)')
    plt.title('Percentage of Each Pronoun in Tweets')
    plt.xticks(rotation=45)
    plt.show()

if __name__ == "__main__":
    file_path = 'A2task2.1-part-00000.csv'
    
    data = read_and_process_data(file_path)
    
    plot_pronoun_percentages(data)
```
